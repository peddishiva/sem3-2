```python
import nltk
from nltk.tag import hmm
from nltk.corpus import treebank
from nltk.probability import LidstoneProbDist
```

```python
#Download requierd resources
nltk.download('treebank')
nltk.download('universal_tagset')
```

```python
#load Treebank corpus with universal tagset
data = treebank.tagged_sents(tagset='universal')
```

```python
#split data into training and testing
train_data =data[:3500]
test_data =data[3500:]
```

```python
#Initialize HMM trainer
trainer = hmm.HiddenMarkovModelTrainer()
```

```python
#Train HMM with smoothing(important!)
hmm_tagger = trainer.train_supervised(
    train_data,
    estimator = lambda fd,bins:LidstoneProbDist(fd, 0.1, bins)
)
```

```python
#Input sentence(lowercase for better matching)
sentence = "the cat sat on the mat".split()
```

```python
#Tag the sentence
tags = hmm_tagger.tag(sentence)
```

```python
print("Tagged Sentence:")
print(tags)
```
